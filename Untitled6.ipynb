```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
cars_df=pd.read_csv('mtcars.csv')
print(cars_df)
a=plt.scatter(cars_df['mpg'],cars_df['wt'],marker='*')
print(a)
plt.xlabel('MPG')
plt.ylabel('Weight')
plt.title('Relationship between the height of the cars and mpg')
plt.show()
trans=cars_df['am'].value_counts()
print(trans)
trans.plot.bar()
plt.xlabel("transmission")
plt.title("Barplottocheckthefrequencydistributionoftransmissiontyp")
plt.show()
cars_df['mpg'].plot(kind='box')
plt.xlabel('MPG')
plt.title('Boxandwhiskerplotofmpg')
plt.show()
minvalue=np.min(cars_df['mpg'])
q1=np.percentile(cars_df['mpg'],25)
median=np.median(cars_df['mpg'])
q3=np.percentile(cars_df['mpg'],75)
maxvalue=np.max(cars_df['mpg'])
print('minimumvalue:',minvalue)
print('Q1:',q1)
print('median:',median)
print('Q3:',q3)
print('maxvalue:',maxvalue)
```
