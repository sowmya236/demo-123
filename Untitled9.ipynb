```python
import pandas as pd
import numpy as np
Series_A = pd.Series([10, 20, 30, 40, 50])
Series_B = pd.Series([40, 50, 60, 70, 80])
union = pd.Series(np.union1d(Series_A.values, Series_B.values))
intersect = pd.Series(np.intersect1d(Series_A.values, Series_B.values))
notcommanSeries=union[-union.isin(intersect)]
print(notcommanSeries)
print("largest in Series_A is:",Series_A.max())
print("smallest number in Series_A is:",Series_A.min())
print("the sum of Series_B is:",Series_B.sum())
print("the  avarage of Series_A is:",Series_A.mean())
print("the median of Series_B is :",Series_B.median())
```
